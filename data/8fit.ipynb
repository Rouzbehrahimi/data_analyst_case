```python
new_df = pd.merge(subscriptions, spendings,  how='left', left_on=['signup_country_code','marketing_channel_id','signup_platform','subscription_date'], right_on = 
                  ['country_code','marketing_channel_id','platform','report_date'])
```

```python
df=new_df.dropna()
```

```python
platform=df.groupby('signup_platform')['spendings','net_revenue','subscription_count'].sum()
```

```python
platform.drop('Spending/Rev',axis=1,inplace=True)
```

```python
platform['CAC']=platform['spendings']/platform['subscription_count']
platform['MarketingCost/Revenue']=platform['spendings']/platform['net_revenue']
platform['Marketin-Revenue']=(platform['net_revenue']-platform['spendings'])/platform['net_revenue']
platform['UnnitPrice']=platform['net_revenue']/platform['subscription_count']
platform
```

```python
df['costumer_COA']=df['spendings']/df['subscription_count']
```

```python
Marketing=df.groupby('marketing_channel_id')['spendings','net_revenue','subscription_count'].sum()
```

```python
Marketing['CAC']=Marketing['spendings']/Marketing['subscription_count']
Marketing['spendings/revenue']=Marketing['spendings']/Marketing['net_revenue']
Marketing.sort_values('spendings',ascending=False)
```
